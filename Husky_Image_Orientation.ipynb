```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing OpenCv library
import cv2
```

```python
img = cv2.imread('husky.jpg')
```

```python
plt.imshow(img)
```

# Here one thing to notice is that OpenCV by default has BGR channel not RGB, hence we saw this wiered kind of image, however matplotlib is configured by default as RGB channel

```python
type(img)
```

```python
img.shape
```

```python
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

## Converting this into grayscale

```python
gray_scale = cv2.imread('husky.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_scale,cmap='gray')
```

```python
plt.imshow(rgb)
```

## Using Image ratio (Alternative)

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(rgb,(0,0),rgb,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
img_flip = cv2.flip(rgb,1)
```

```python
plt.imshow(img_flip)
```
